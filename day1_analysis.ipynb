## DATA VISUALIZATION

```python
# Import libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Category_sales = df.groupby('Category')['Sales'].sum()

Category_sales.plot(kind='bar')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()

```

## Sales by Category

- Visualised total Sales by categories 
- Helps identify top revenue-generating Categories


# Profit by Category

```python
Category_Profit = df.groupby('Category')['Profit'].sum()

Category_Profit.plot(kind='bar')
plt.title("Profit by Categories")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.show()
```

## Profit by Category

- Compared profits across categories
- Helps identify most profitable category
- observed that despite having a high sale furniture category has low profits 

# Profit by sub category

```python
subcat_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values()

subcat_profit.plot(kind='barh', figsize=(8,6))
plt.title("Profit by Sub Categories")
plt.xlabel("Profit")
plt.ylabel("Sub-Categories")
plt.show
```

## Profit by Sub-Categories

- Identified the Loss-making and the high-profit making sub-categories
- Helps in making product level decisions

# Discounts vs Profit

```python
sns.scatterplot(x='Discount',y = 'Profit', data = df)
plt.title("Discounts vs Profit")
plt.show()
```

# Region wise analysis

```python
region_sales = df.groupby('Region')['Sales'].sum()

region_sales.plot(kind='bar',color='r')
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

## Sales by Region

- Compared Sales of different Regions 
- Helps Identify the stronger and the weaker markets

# Profit Distribution

```python
sns.histplot(df['Profit'], bins=50)
plt.title("Profit distribution")
plt.show()
```

## 📊 Key Observations from Visuals

- Technology category stands out as the most profitable

- Sub-categories like Tables, Bookcases, and Supplies are showing losses and need attention

- Higher discounts are negatively impacting profit
